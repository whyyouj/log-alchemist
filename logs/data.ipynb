```python
import pandas as pd

df_cc = pd.read_csv("Credit card transactions - India - Simple.csv")
df_cc
```

```python
import plotly.express as px
card_type_counts = df_cc["Card Type"].value_counts()

# Convert the Series to a DataFrame for easier plotting
card_type_counts_df = card_type_counts.reset_index()
card_type_counts_df.columns = ['Card Type', 'Count']

# Create a bar chart with Plotly
fig = px.bar(card_type_counts_df, x='Card Type', y='Count', title='Card Type Distribution', 
             labels={'Card Type': 'Card Type', 'Count': 'Number of Cards'})

# Show the figure
fig.write_image("card_type_distribution.png")
```
